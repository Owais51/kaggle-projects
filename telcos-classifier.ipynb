```python
y=data.Churn
y
```

```python
x=data.drop(columns=['Churn','customerID','TotalCharges'])
x
```

```python
x.dtypes
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
```

```python
x_train.shape
```

```python
x.head(5)
```

```python
x.PaperlessBilling.unique()
```

```python
nom_cols=[6,7,8,9,10,11,12,13,14,16]
ord_cols=[0,2,3,5,15]
```

```python
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
from sklearn.compose import make_column_transformer
from sklearn import set_config
trans=make_column_transformer((OneHotEncoder(sparse='false'),nom_cols),(OrdinalEncoder(),ord_cols),remainder='passthrough')
set_config(display='diagram')
trans
```

```python
from sklearn.neighbors import KNeighborsClassifier
algo=KNeighborsClassifier(n_neighbors=13)
algo
```

```python
from sklearn.pipeline import make_pipeline
pipe=make_pipeline(trans,algo)
pipe
```

```python
pipe.fit(x_train,y_train)
```

```python
pred=pipe.predict(x_test)
pred
```

```python
x.info()
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(pred,y_test)*100
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import VotingClassifier
```

```python
algorithm1=LogisticRegression(solver='liblinear')
algo=KNeighborsClassifier()
algorithm3=DecisionTreeClassifier()
```

```python
vote=VotingClassifier(estimators=[('x1',algorithm1),('x2',algo),('x3',algorithm3)],voting='soft',verbose=True)
vote
```

```python
pipe_1=make_pipeline(trans,vote)
pipe_1
```

```python
pipe_1.fit(x_train,y_train)
```

```python
pred_1=pipe_1.predict(x_test)
pred_1
```

```python
accuracy_score(pred_1,y_test)*100
```
