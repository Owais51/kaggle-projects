```python
nom_cols=[0,3,11,12,13,14,18,19]
ord_cols=[11,21,25,29]
```

```python
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
from sklearn.compose import make_column_transformer
from sklearn import set_config
trans=make_column_transformer((OneHotEncoder(sparse='false'),nom_cols),(OrdinalEncoder(),ord_cols),remainder='passthrough')
set_config(display='diagram')
trans
```

```python
from sklearn.neighbors import KNeighborsClassifier
algo=KNeighborsClassifier(n_neighbors=13)
algo
```

```python
from sklearn.pipeline import make_pipeline
pipe=make_pipeline(trans,algo)
pipe
```

```python
pipe.fit(x_train,y_train)
```

```python
pred=pipe.predict(x_test)
pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(pred,y_test)*100
```
